```r
#setwd("/Users/Adarsh/Desktop/") # Please set the working directory to where the jellyfish.csv file is stored 
install.packages("quantmod")
library(quantmod)  # Install the library quantmod if not there

data= read.csv("jellyfish.csv")


Read_ipi<-function(y1,x1)
{

y_peak=findPeaks(y1,thresh=0)

time_value=c()

print(y_peak)

for(i in seq((length(y_peak)-1)))
  time_value=c(time_value,((y_peak[i+1]-y_peak[i])*0.066666667))



time_final=c()
for (i in seq(length(time_value))) {
      if(time_value[i]>0.5)
        time_final=c(time_final,time_value[i])
}
return (time_final)

}

# For Jellyfish 1
y1 <-data$Jellyfish.1.Area..cm.2.
x1 <-data$time..s.

time_final_day<-Read_ipi(y1[1:4001],x1[1:4001]) # Day
time_final_night<-Read_ipi(y1[4002:8001],x1[4002:8001]) 


p1=hist(time_final_day,xlab = "IPI",main = "Histogram of Jellyfish 1",col = c("grey"))
p2=hist(time_final_night,xlab = "IPI",main = "Histogram of Jellyfish 1")
plot( p1, col=rgb(0,0,1,1/4), xlim=c(0,3),main = "Jellyfish 1")  # histogram for day
plot( p2, col=rgb(1,0,0,1/4), xlim=c(0,3),main = "Jellyfish 1", add=T)  # histogram for night
write.csv(time_final_day, file = "jellyfish1_IPI_day.csv")
write.csv(time_final_night, file = "jellyfish1_IPI_night.csv")


# For Jellyfish 2
y1 <-data$Jellyfish.2.Area..cm.2.
x1 <-data$time..s.

time_final_day<-Read_ipi(y1[1:4001],x1[1:4001]) # Day
time_final_night<-Read_ipi(y1[4002:8001],x1[4002:8001]) 


p1=hist(time_final_day,xlab = "IPI",main = "Histogram of Jellyfish 2 day",col = c("grey"))
p2=hist(time_final_night,xlab = "IPI",main = "Histogram of Jellyfish 2 night")
plot( p1, col=rgb(0,0,1,1/4), xlim=c(0,3),main = "Jellyfish 2")  # histogram for day
plot( p2, col=rgb(1,0,0,1/4), xlim=c(0,3),main = "Jellyfish 2", add=T)  # histogram for night

write.csv(time_final_day, file = "jellyfish2_IPI_day.csv")
write.csv(time_final_night, file = "jellyfish2_IPI_night.csv")


# For Jellyfish 3
y1 <-data$Jellyfish.3.Area..cm.2.
x1 <-data$time..s.

time_final_day<-Read_ipi(y1[1:4001],x1[1:4001]) # Day
time_final_night<-Read_ipi(y1[4002:8001],x1[4002:8001]) 


p1=hist(time_final_day,xlab = "IPI",main = "Histogram of Jellyfish 3 day",col = c("grey"))
p2=hist(time_final_night,xlab = "IPI",main = "Histogram of Jellyfish 3 night")
plot( p1, col=rgb(0,0,1,1/4), xlim=c(0,3),main = "Jellyfish 3")  # histogram for day
plot( p2, col=rgb(1,0,0,1/4), xlim=c(0,3),main = "Jellyfish 3", add=T)  # histogram for night

write.csv(time_final_day, file = "jellyfish3_IPI_day.csv")
write.csv(time_final_night, file = "jellyfish3_IPI_night.csv")

# For Jellyfish 4
y1 <-data$Jellyfish.4.Area..cm.2.
x1 <-data$time..s.

time_final_day<-Read_ipi(y1[1:4001],x1[1:4001]) # Day
time_final_night<-Read_ipi(y1[4002:8001],x1[4002:8001]) 


p1=hist(time_final_day,xlab = "IPI",main = "Histogram of Jellyfish 4 day",col = c("grey"))
p2=hist(time_final_night,xlab = "IPI",main = "Histogram of Jellyfish 4 night")
plot( p1, col=rgb(0,0,1,1/4), xlim=c(0,3),main = "Jellyfish 4")  # histogram for day
plot( p2, col=rgb(1,0,0,1/4), xlim=c(0,3),main = "Jellyfish 4", add=T)  # histogram for night

write.csv(time_final_day, file = "jellyfish4_IPI_day.csv")
write.csv(time_final_night, file = "jellyfish4_IPI_night.csv")

# For Jellyfish 5
y1 <-data$Jellyfish.5.Area..cm.2.
x1 <-data$time..s.

time_final_day<-Read_ipi(y1[1:4001],x1[1:4001]) # Day
time_final_night<-Read_ipi(y1[4002:8001],x1[4002:8001]) 


p1=hist(time_final_day,xlab = "IPI",main = "Histogram of Jellyfish 5 day",col = c("grey"))
p2=hist(time_final_night,xlab = "IPI",main = "Histogram of Jellyfish 5 night")
plot( p1, col=rgb(0,0,1,1/4), xlim=c(0,3),main = "Jellyfish 5")  # histogram for day
plot( p2, col=rgb(1,0,0,1/4), xlim=c(0,3),main = "Jellyfish 5", add=T)  # histogram for night

write.csv(time_final_day, file = "jellyfish5_IPI_day.csv")
write.csv(time_final_night, file = "jellyfish5_IPI_night.csv")

# For Jellyfish 6
y1 <-data$Jellyfish.6.Area..cm.2.
x1 <-data$time..s.

time_final_day<-Read_ipi(y1[1:4001],x1[1:4001]) # Day
time_final_night<-Read_ipi(y1[4002:8001],x1[4002:8001]) 




p1=hist(time_final_day,xlab = "IPI",main = "Histogram of Jellyfish 6 day",col = c("grey"))
p2=hist(time_final_night,xlab = "IPI",main = "Histogram of Jellyfish 6 night")
plot( p1, col=rgb(0,0,1,1/4), xlim=c(0,3),main = "Jellyfish 6")  # histogram for day
plot( p2, col=rgb(1,0,0,1/4), xlim=c(0,3),main = "Jellyfish 6", add=T)  # histogram for night

write.csv(time_final_day, file = "jellyfish6_IPI_day.csv")
write.csv(time_final_night, file = "jellyfish6_IPI_night.csv")

# For Jellyfish 7
y1 <-data$Jellyfish.7.Area..cm.2.
x1 <-data$time..s.

time_final_day<-Read_ipi(y1[1:4001],x1[1:4001]) # Day
time_final_night<-Read_ipi(y1[4002:8001],x1[4002:8001]) 


p1=hist(time_final_day,xlab = "IPI",main = "Histogram of Jellyfish 7 day",col = c("grey"))
p2=hist(time_final_night,xlab = "IPI",main = "Histogram of Jellyfish 7 night")
plot( p1, col=rgb(0,0,1,1/4), xlim=c(0,3),main = "Jellyfish 7")  # histogram for day
plot( p2, col=rgb(1,0,0,1/4), xlim=c(0,3),main = "Jellyfish 7", add=T)  # histogram for night

write.csv(time_final_day, file = "jellyfish7_IPI_day.csv")
write.csv(time_final_night, file = "jellyfish7_IPI_night.csv")

# For Jellyfish 8
y1 <-data$Jellyfish.8.Area..cm.2.
x1 <-data$time..s.

time_final_day<-Read_ipi(y1[1:4001],x1[1:4001]) # Day
time_final_night<-Read_ipi(y1[4002:8001],x1[4002:8001]) 


p1=hist(time_final_day,xlab = "IPI",main = "Histogram of Jellyfish 8 day",col = c("grey"))
p2=hist(time_final_night,xlab = "IPI",main = "Histogram of Jellyfish 8 night")
plot( p1, col=rgb(0,0,1,1/4), xlim=c(0,3),main = "Jellyfish 8")  # histogram for day
plot( p2, col=rgb(1,0,0,1/4), xlim=c(0,3),main = "Jellyfish 8", add=T)  # histogram for night

write.csv(time_final_day, file = "jellyfish8_IPI_day.csv")
write.csv(time_final_night, file = "jellyfish8_IPI_night.csv")

lag.plot(y1)
```

```r
# Scatterplot of mpg vs. hp for each combination of gears and cylinders
# in each facet, transmittion type is represented by shape and color
qplot(hp, mpg, data=mtcars, shape=am, color=am, 
   facets=gear~cyl, size=I(3),
   xlab="Horsepower", ylab="Miles per Gallon") 
```

```r
# Separate regressions of mpg on weight for each number of cylinders
qplot(wt, mpg, data=mtcars, geom=c("point", "smooth"), 
   method="lm", formula=y~x, color=cyl, 
   main="Regression of MPG on Weight", 
   xlab="Weight", ylab="Miles per Gallon")
```

```r
# Boxplots of mpg by number of gears 
# observations (points) are overlayed and jittered
qplot(gear, mpg, data=mtcars, geom=c("boxplot", "jitter"), 
   fill=gear, main="Mileage by Gear Number",
   xlab="", ylab="Miles per Gallon")
```
